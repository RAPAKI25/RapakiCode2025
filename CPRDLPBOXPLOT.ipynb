```python
import matplotlib.pyplot as plt
import numpy as np

# Data
surface_types = [
    "Smooth Pāhoehoe",
    "Hummocky Pāhoehoe",
    "Rubbly Pāhoehoe",
    "Blocky to High Relief 'a'ā",
    "Blocky"
]

cpr_values = [0.34, 0.48, 0.73, 0.69, 1.14]
cpr_std = [0.11, 0.19, 0.24, 0.25, 0.32]

dlp_values = [0.18, 0.2, 0.19, 0.2, 0.19]
dlp_std = [0.05, 0.06, 0.05, 0.06, 0.05]

# Creating the figure and axes
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)

# Define colors and markers
colors = ['blue', 'black']  # Green and Purple
markers = ['o', 's']
error_bar_color = 'red'

# Plot CPR with error bars
ax.errorbar(
    range(1, len(cpr_values) + 1), cpr_values, yerr=cpr_std, fmt=markers[0],
    label='CPR', capsize=5, capthick=2, elinewidth=1.5, markersize=8, color=colors[0], ecolor=error_bar_color, alpha=0.8
)

# Plot DLP with error bars
ax.errorbar(
    range(1, len(dlp_values) + 1), dlp_values, yerr=dlp_std, fmt=markers[1],
    label='DLP', capsize=5, capthick=2, elinewidth=1.5, markersize=8, color=colors[1], ecolor=error_bar_color, alpha=0.8
)

# Formatting
plt.rcParams['font.family'] = 'Times New Roman'
ax.set_xticks(range(1, len(surface_types) + 1))
ax.set_xticklabels(surface_types, rotation=30, ha='right', fontsize=12)
ax.set_title("CPR and DLP with Error", fontsize=14, fontweight='bold')
ax.set_ylabel("Measured Values", fontsize=12)
ax.set_xlabel("Surface Roughness Types", fontsize=12)
ax.legend(fontsize=12, frameon=True, loc='upper left', edgecolor='black')
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Save or show the figure
plt.tight_layout()
plt.show()
```
