```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Sample data
np.random.seed(42)
data = pd.DataFrame({
    'VV': np.random.uniform(-10, 10, 40),  # Simulated VV backscatter
    'VH': np.random.uniform(-10, 10, 40),  # Simulated VH backscatter
})
data['Polarization_Ratio'] = data['VV'] / data['VH']  # Polarization ratio

# Create the scatter plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    data['VV'], 
    data['VH'], 
    c=data['Polarization_Ratio'], 
    cmap='viridis', 
    edgecolor='k'
)
plt.colorbar(scatter, label='Polarization Ratio (VV/VH)')
plt.xlabel('VV Polarization')
plt.ylabel('VH Polarization')
plt.title('Scatter Plot of VV vs. VH with Polarization Ratio')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate data
np.random.seed(42)
n_samples = 100
crystallinity = np.random.uniform(0, 1, n_samples)  # Crystallinity index (0 = low, 1 = high)
reflectance = np.random.uniform(0.1, 0.5, n_samples) + 0.3 * crystallinity  # Reflectance affected by crystallinity

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    crystallinity, 
    reflectance, 
    c=crystallinity, 
    cmap='Greys_r',  # '_r' reverses the 'Greys' colormap, making 0 black and 1 white
    edgecolor='k', 
    s=80, 
    label='Basaltic Rocks'
)

# Add a colorbar
colorbar = plt.colorbar(scatter)
colorbar.set_label('Crystallinity Index (0 = Low, 1 = High)')

# Label the plot
plt.xlabel('Crystallinity Index (0 = Low, 1 = High)')
plt.ylabel('Reflectance')
plt.title('Crystallinity and Reflectance Relationship in Basaltic Rocks')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
